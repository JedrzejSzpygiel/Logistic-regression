# Logistic and softmax regression on Iris dataset

Loading data from sklearn datasets

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
iris = datasets.load_iris()
X = iris["data"][:,3:] #petal width
y = (iris["target"]==2).astype(int)
list(iris.keys())
```

Logistic refression model import and training

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X,y);
```

Plotting the decision boundary for Iris-Virginica and the rest of species

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:,1], "g-", label = "Iris-Virginica");
plt.plot(X_new, y_proba[:,0], "b--", label = "Rest");
plt.legend();
```

Applying softmax regression

```python
X = iris["data"][:,(2,3)] #petal length and width
y = iris["target"]

softmax_reg = LogisticRegression(multi_class="multinomial", solver = "lbfgs", C = 10)
softmax_reg.fit(X,y)
```
